###   UNIVARIATE  ###

```python
mean = data['Age'].mean()
median = data['Age'].median()
fig, axes = plt.subplots(1,2, figsize = (16,6))
data['Age'].plot(kind='hist', ax = axes[0])
data['Age'].plot(kind='density', ax = axes[1])
plt.axvline(mean,color='r',label='Mean')   
plt.axvline(median,color='b',label='Median')
plt.legend()  
plt.show()
```

```python
sns.boxplot(data['Age'])
plt.show()
```

```python
# Division Name
fig, axes = plt.subplots(1,2, figsize = (15,5))
brk= (0.1, 0.1, 0.1)
data['Division Name'].value_counts().plot(kind='pie',autopct = '%.1f%%',explode=brk,ax = axes[0]) 
data['Division Name'].value_counts().plot(kind = 'bar', ax = axes[1])
plt.show()

# In the General divison -> people wrote most review ie 59% (> 12000)
```

```python
# Department Name
fig, axes = plt.subplots(1,2, figsize = (15,5))
brk= (0.1, 0.1, 0.1,.1,.1,.1)
data['Department Name'].value_counts().plot(kind='pie',autopct = '%.1f%%',explode=brk,ax = axes[0]) 
data['Department Name'].value_counts().plot(kind = 'bar', ax = axes[1])
plt.show()

# In the Tops Department  -> people wrote most review ie 44% (> 10000)
```

```python
# Class Name
fig, axes = plt.subplots(1,2, figsize = (19,8))
data['Class Name'].value_counts().plot(kind='pie',autopct = '%.1f%%',ax = axes[0]) 
data['Class Name'].value_counts().plot(kind = 'bar', ax = axes[1])
plt.show()

# In the Dress Class  -> people wrote most review ie 26% (> 12000)
```

```python
# Rating
fig, axes = plt.subplots(1,2, figsize = (19,8))
data['Rating'].value_counts().plot(kind='pie',autopct = '%.1f%%',ax = axes[0]) 
data['Rating'].value_counts().plot(kind = 'bar', ax = axes[1])
plt.show()
 
# No of people rated 5 which is more than 12000 ie by 55.9%
```
